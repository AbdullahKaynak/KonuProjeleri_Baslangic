```python
GivenString = "Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
```

```python
class TextAnalyzer(object):
    
    def __init__(self,text):
        ## Noktalama işaretlerini kaldır, Metni Küçük harfe çevir
        formattedText = text.replace('.','').replace('!','').replace('?','').replace(',','').lower()

        ## Biçimlendirilmiş metni fmtText'e adında biryere ata
        self.fmtText = formattedText 

    def freqAll(self):
        ## split ile tek tek ayur
        wordList = self.fmtText.split(" ")

        ## boş sözlük oluştur adı freqMap olsun
        freqMap = {}

        ## Listedeki yinelenenleri kaldırmak için `set` işlevini kullanın.
        for word in set(wordList):
            freqMap[word] = wordList.count(word)
        return freqMap

    def freqOf(self,word):
        ## bulunması istenilen kelime (Frekans haritasını al)
        freqDict = self.freqAll()

        if word in freqDict:
            return freqDict[word]
        else:
            return 0
```

# Adım 1: TextAnalyzer sınıfının bir örneğini oluşturun.

```python
analiz = TextAnalyzer(GivenString)
```

# Adım 2: Verileri küçük harfe dönüştüren fonksiyonu çağırın

```python
print(analiz.fmtText)
```

# Adım 3: Verilerdeki tüm benzersiz kelimelerin sıklığını sayan fonksiyonu çağırın.
