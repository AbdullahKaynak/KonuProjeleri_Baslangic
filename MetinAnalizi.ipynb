### Görev 1. Metni küçük harfe dönüştürme: Müşteri geri bildirimlerini ön işlemden geçirerek tüm metni küçük harfe dönüştürmek istiyorsunuz. Bu adım, metni standartlaştırmaya yardımcı olur. Metni küçük harfe dönüştürmek, belirli harf büyük/küçük harf kullanımına değil, içeriğe odaklanmanızı sağlar.

### Görev 2. Belirli bir metin dizesindeki tüm kelimelerin sıklığı: Metni küçük harfe dönüştürdükten sonra, müşteri geri bildirimlerindeki her kelimenin sıklığını belirlemek istiyorsunuz. Bu bilgi, hangi kelimelerin daha sık kullanıldığını belirlemenize yardımcı olacak ve müşterilerin yorumlarında bahsettikleri temel yönleri veya konuları gösterecektir. Kelime sıklıklarını analiz ederek, müşteriler tarafından dile getirilen en yaygın sorunlar hakkında fikir edinebilirsiniz.

### Görev 3. Belirli bir kelimenin sıklığı: Genel kelime sıklıklarını analiz etmenin yanı sıra, analizinizle ilgili belirli bir kelimenin sıklığını da özel olarak izlemek istiyorsunuz. Örneğin, ürünün güvenilirliği hakkındaki müşteri görüşlerini ölçmek için müşteri yorumlarında "güvenilir" kelimesinin ne sıklıkla geçtiğini izlemek isteyebilirsiniz. Belirli bir kelimenin sıklığına odaklanarak, o özel yönle ilgili müşteri görüşleri veya tercihleri ​​hakkında daha derin bir anlayış kazanabilirsiniz.

## Adım 1: Bir dize tanımlayın¶

```python
GivenString = "Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
```

### Senaryoda belirtilen görevleri gerçekleştirmek için üç farklı metoda sahip bir sınıf oluşturmanız gerekmektedir.

## Adım 2: Sınıfı ve özelliklerini tanımlayın.
#### 1- TextAnalyzer adında bir sınıf oluştur
#### 2- Metin Argümanı alan __init__ kurucu yöntemini tanımla

## Adım 3 - Metni küçük harfe dönüştürmek için bir kod uygulayın
#### 1- Yapıcı fonksiyonun içinde, lower() metodunu kullanarak metin argümanını küçük harfe dönüştürün.
#### 2- Ardından, replace() metodunu kullanarak metinden noktalama işaretlerini (nokta, ünlem işareti, virgül ve soru işareti) kaldırın.
#### 3- Son olarak, biçimlendirilmiş metni fmtText adlı yeni bir özniteliğe atayın.

## Adım 4 : Tüm benzersiz kelimelerin sıklığını saymak için bir kod uygulayın
### Bu adımda freqAll() yöntemini aşağıdaki parametrelerle uygulayacaksınız:
#### 1 - fmtText niteliğini split() yöntemini kullanarak tek tek kelimelere bölün.
#### 2 - Kelime sıklığını saklamak için boş bir sözlük oluşturun.
#### 3 - Kelime listesini yineleyin ve sıklık sözlüğünü buna göre güncelleyin.
#### 4 -  Oluşumu saymak için sayım yöntemini kullanın.
#### 5 - Frekans sözlüğünü döndürün.

## Adım 5: Belirli bir kelimenin sıklığını saymak için bir kod uygulayın.
### 5. adımda, kelime argümanı alan freqOf(word) yöntemini uygulamanız gerekiyor:
#### 1-Bir yöntem oluşturun ve bulunması gereken kelimeyi bu yönteme iletin.
#### 2-Kelimenin listede olup olmadığını kontrol etmek için freqAll yöntemini kullanın.
#### 3- Kelimenin sayısını döndürün. Kelime bulunamazsa, döndürülen sayı 0 olur.

```python
# Metni analiz etmek için TextAnalyzer adında bir sınıf oluşturalım.
class TextAnalyzer(object):
    def __init__(self, text):
        formattedText = text.replace(".", "").replace("!", "").replace(",", "").replace("?", "")
        formattedText = formattedText.lower()

        self.fmtText = formattedText

    def fireqAll(self):
        kelimeler = self.fmtText.split(" ")
        
        freqMap = {}
        for word in set(kelimeler):
            freqMap[word] = kelimeler.count(word)
        return freqMap
    
    def fireq0f(self, word):
        freqDict = self.fireqAll()

        if word in freqDict:
            return freqDict[word]
        else:
            return 0
```

## Adım 1: TextAnalyzer sınıfının bir örneğini oluşturun.
Verilen dizeyi argüman olarak geçirerek TextAnalyzer sınıfının bir örneğini oluşturun.

```python
analyzed = TextAnalyzer(GivenString)
```

## Adım 2: Verileri küçük harfe dönüştüren fonksiyonu çağırın.

```python
print("Yenilenmiş yazı: ", analyzed.fmtText)
```

## Adım 3: Verilerdeki tüm benzersiz kelimelerin sıklığını sayan fonksiyonu çağırın.

```python
analyzed.fireqAll()
```

## Adım 4: Belirli bir kelimenin sıklığını sayan fonksiyonu çağırın.
Burada, "lorem" kelimesinin sıklığını sayan fonksiyonu çağıracaksınız.

```python
word = "lorem"
frequency = analyzed.fireq0f(word)
print(f"İstenen kelime olan {word} 'den toplam: {frequency} tane vardır.")
```
