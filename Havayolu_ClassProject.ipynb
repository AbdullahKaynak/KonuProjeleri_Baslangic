```python
class Ucus():
    havayolu = "Pegasus Airlines"

    def __init__(self, kod, kalkis_yeri, varis_yeri, ucus_suresi, bilet_fiyati, yolcu_kapasitesi, yolcu):
        self.kod = kod
        self.kalkis_yeri = kalkis_yeri
        self.varis_yeri = varis_yeri
        self.ucus_suresi = ucus_suresi
        self.bilet_fiyati = bilet_fiyati
        self.yolcu_kapasitesi = yolcu_kapasitesi    
        self.yolcu = yolcu
    
    def anons_yap(self):
        return "{} sefer sayılı {}-{} uçuşumuz yaklaşık {} dakika sürecektir. ".format(
            self.kod,
            self.kalkis_yeri,
            self.varis_yeri,
            self.ucus_suresi
        )

    def koltuk_sayisi_guncelle(self):
        return self.yolcu_kapasitesi - self.yolcu
        

    def bilet_satis(self, bilet_Adedi=1 ):
        if self.yolcu + bilet_Adedi <= self.yolcu_kapasitesi:
            self.yolcu += bilet_Adedi
            self.koltuk_sayisi_guncelle() 
            print("{} adet bilet satılmıştır. kalan koltuk sayısı: {}". format(
            bilet_Adedi,
            self.koltuk_sayisi_guncelle())) 
        else:
            print("Yeterli koltuk bulunmamaktadır. Kalan koltuk sayısı: {}". format(
            self.koltuk_sayisi_guncelle()))
    
    def bilet_iptal(self, bilet_adeti = 1):
        if self.yolcu >= bilet_adeti:
            self.yolcu -= bilet_adeti
            print("{} adet bilet iptal edilmiştir. Kalan koltuk sayısı: {}". format(
            bilet_adeti,
            self.koltuk_sayisi_guncelle()))
        else:
            print("İptal edilecek bilet sayısı mevcut yolcu sayısından fazla olamaz. Kalan koltuk sayısı: {}". format(
            self.koltuk_sayisi_guncelle()))

    def ciro_hesapla(self):
        return self.yolcu * self.bilet_fiyati
    

        
```

```python
ucak1 = Ucus("PC123", "Istanbul", "Izmir", 60, 1000, 180, 150)
```

### Sorgu: ucak1 nesnesi için şu an boşta kaç koltuk var? (Metodunu çağır).

```python
ucak1.yolcu_kapasitesi
```

### Bu uçağa 40 adet bilet sat. (Kalan koltuk sayısını kontrol et).

```python
ucak1.bilet_satis(40)
```

### Satıştan sonra 10 adet bilet iptal et.

```python
ucak1.bilet_iptal(10)
```

### Son durumda uçaktaki güncel yolcu sayısı nedir ve boş koltuk kaç kaldı?

```python
ucak1.yolcu
ucak1.koltuk_sayisi_guncelle()
```
